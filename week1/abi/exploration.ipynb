```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../dataset/comtrade.csv')
```

```python
df.shape
```

```python
df.describe()
```

```python
corr = df.corr()
corr[corr > 0.4 ]
```

```python
df.skew()
```

```python
grouped_countries = df.groupby(['Reporter', 'Trade Flow']).agg({'Trade Value (US$)': ['count', 'mean', 'min', 'max', 'sum','std'] })
grouped_countries.columns = ['count', 'mean (US$)', 'min (US$)', 'max (US$)', 'sum (US$)', 'std',]
grouped_countries = grouped_countries.reset_index()
print(grouped_countries)
```

```python
grouped_countries.plot(kind='line', subplots=True, layout=(3,3), figsize=(10,7));
```

```python
plt.figure(figsize=(10,7))
sns.barplot('Reporter', 'Trade Value (US$)', data=df[df['Reporter']=='Armenia'])
```
